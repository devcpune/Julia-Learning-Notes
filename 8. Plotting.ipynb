# Plotting

There are multiple ways to create a plot in Julia, we can use packages other languages(like PyPlot). Or we can use packages available in Julia itself.

Let's start by using `Plots.jl`.

We first need to install `Plots`:

```julia
#Pkg.add("Plots")
using Plots
```

`Plots.jl` allows you to seemlessly change backend. Here we'll try out the `gr()` and  `unicodeplots()`  backends.

Let's generate some dummy data to plot!

```julia
x = -3:0.3:3
cubes(n) = n^3

y = cubes.(x)
```

Now let's use GR backend

```julia
gr()
```

```julia
plot(x, y, label="line" )
scatter!(x, y, label="points")
```

You may have noticed here `scatter` is followed by a `!` sign that is because we modifying the existing plot instead of creating a new one. Try playing around with the `!` sign to get a better understanding.

Now we can use any other backend instead of `g

```julia
unicodeplots()
```

```julia
plot(x, y, label="line" )
scatter!(x, y, label="points")
```

```julia
gr()
```
